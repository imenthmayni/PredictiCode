```python
import pandas as pd
import json

# Charger les données des fichiers JSON
with open('issues.json') as f:
    issues = json.load(f)

with open('commits.json') as f:
    commits = json.load(f)

# Normaliser les données JSON en DataFrame
df_issues = pd.json_normalize(issues)
df_commits = pd.json_normalize(commits)

```

```python
df_commits.rename(columns={'Message': 'Commit Message'}, inplace=True)
```

```python
df_commits.drop(columns=['Modified Files'], inplace=True)
```

```python
df_issues.rename(columns={'number': 'IdTicket'}, inplace=True)
df_issues.drop(columns=['title'], inplace=True)
```

```python
df_commits['IdTicket'] = df_commits['IdTicket'].astype(str)
df_issues['IdTicket'] = df_issues['IdTicket'].astype(str)

dataset = pd.merge(df_commits, df_issues, on='IdTicket', how='left')
```

```python
df = pd.read_csv('functions_bug.csv')
df.head()
```

```python
for col in df.columns[2:]:
    # Remplacer NaN par 0 et "modified" par 1
    df[col] = df[col].fillna(0).apply(lambda x: 1 if x == 'modified' else x)
    df[col] = df[col].astype(int)

df
```

```python
# Define the columns that represent functions
function_columns = df.columns[2:]
```

```python
# Create the 'functions' column
df['functions'] = df.apply(lambda row: [col for col in function_columns if row[col] == 1], axis=1)
```

```python
# Drop the original function columns 
df = df.drop(columns=function_columns)
df 
```

```python
df.drop(columns=['message'], inplace=True)
df.rename(columns={'commit': 'IdCommit'}, inplace=True)
```

```python
df['IdCommit'] = df['IdCommit'].astype(str)
dataset['IdCommit'] = dataset['IdCommit'].astype(str)
dataset = pd.merge(dataset, df, on='IdCommit', how='left')
```

```python
# dataset.set_index('IdCommit', inplace=True)
# dataset.head()
```

```python
duplicates = dataset['IdTicket'].duplicated().sum()
duplicates
```

```python
dataset = dataset.drop_duplicates(subset=['IdTicket'], keep='first')
```

```python
duplicates = dataset['IdTicket'].duplicated().sum()
duplicates
```

```python
# Supprimer les lignes où 'IdTicket' est NaN
dataset = dataset[dataset['IdTicket'] != "NaN"]
dataset
```

```python
# Define the desired column order
desired_order = ['IdTicket', 'Date', 'IdCommit', 'Commit Message', 'functions', 'user', 'Author', 'created_at', 'updated_at', 'labels', 'state','duration']
# Reorder the columns 
dataset = dataset.reindex(columns=desired_order)
dataset
```

```python
# Supprimer les lignes où la colonne 'functions' contient des NaN
dataset = dataset[~dataset['functions'].isna()]
dataset
```

```python
print(dataset['labels'].value_counts())
```

```python
# Convertir les listes vides en chaînes vides
dataset['labels'] = dataset['labels'].apply(lambda x: '' if isinstance(x, list) and len(x) == 0 else x)
# Remplacer les chaînes vides par 0
dataset['labels'] = dataset['labels'].replace('', 0)
dataset
```

```python
print(dataset['labels'].value_counts())
```

```python
# Fonction pour convertir des listes en chaînes de caractères
def convert_list_to_string(value):
    if isinstance(value, list):
        return ', '.join(value)
    else:
        return value

# Appliquer la fonction à la colonne 'labels'
dataset['labels'] = dataset['labels'].apply(convert_list_to_string)
```

```python
print(dataset['labels'].value_counts())
```
