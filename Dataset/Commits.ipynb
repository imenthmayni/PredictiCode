```python
import re
from datetime import datetime

# Définir la fonction pour extraire le message du commit et le numéro de ticket
def extract_commit_message(commit_text):
    message_pattern = re.compile(r'\n\n\s*(.*?)\s*\n*\(#(\d+)\)?', re.DOTALL)
    match = message_pattern.search(commit_text)
    if match:
        message = match.group(1).strip()
        ticket_number = match.group(2)
        return message, ticket_number
    else:
        return None, None

# Définir le chemin du fichier
path = r'Desktop/projetGit/datasets/historique_commits.txt'

# Lire le contenu du fichier avec l'encodage latin-1
with open(path, 'r', encoding='latin-1') as file:
    content = file.read()

# Définir les expressions régulières pour extraire les métadonnées des commits
commit_pattern = re.compile(r'commit (\w+)')
author_pattern = re.compile(r'Author: (.+?) <')
date_pattern = re.compile(r'Date:\s+(.+)')
file_pattern = re.compile(r'diff --git a/.* b/(.*)')

# Liste pour stocker les données des commits filtrés
filtered_commits_data = []

# Utiliser finditer pour obtenir tous les matches de commit
for commit_match in commit_pattern.finditer(content):
    start_index = commit_match.start()
    end_index = content.find('commit', start_index + 1)
    if end_index == -1:
        end_index = len(content)
    
    commit_content = content[start_index:end_index].strip()

    # Extraire les métadonnées du commit
    commit_id_match = commit_pattern.search(commit_content)
    commit_id = commit_id_match.group(1) if commit_id_match else 'NaN'

    author_match = author_pattern.search(commit_content)
    author = author_match.group(1) if author_match else 'NaN'

    date_match = date_pattern.search(commit_content)
    date = date_match.group(1) if date_match else 'NaN'

    # Formater la date pour obtenir "YYYY-MM-DD"
    try:
        formatted_date = datetime.strptime(date.strip(), '%a %b %d %H:%M:%S %Y %z').strftime('%Y-%m-%d')
    except ValueError:
        formatted_date = 'NaN'

    files = file_pattern.findall(commit_content)
    files = files if files else 'NaN'

    # Extraire le message du commit et le numéro de ticket
    message, ticket_number = extract_commit_message(commit_content)

    # Vérifier si toutes les clés sont valides (pas de 'N/A')
    if all(key_value != 'NaN' for key_value in (author, formatted_date, message, ticket_number)):
        # Créer un dictionnaire pour stocker les données du commit
        commit_data = {
            'IdCommit': commit_id,
            'Author': author,
            'Date': formatted_date,
            'Modified Files': files,
            'Message': message if message else 'NaN',
            'IdTicket': ticket_number if ticket_number else 'NaN'
        }
        filtered_commits_data.append(commit_data)

# Afficher les commits filtrés qui ne contiennent pas de 'N/A'
for commit in filtered_commits_data:
    print(f"{commit['Author']} {commit['Date']}")

# Afficher les données complètes des commits filtrés
commit_data = []
for commit in filtered_commits_data:
    commit_data.append(commit)

print(commit_data)
```

```python
import json

#Sauvegarde des données dans un fichier JSON
with open('commits.json', 'w') as file:
    json.dump(commit_data, file, indent=4)

print("Commits saved to commits.json")
```
