```python
import requests
from datetime import datetime

# Remplacez par votre token GitHub personnel
token = "your_token"

# Informations sur le dépôt
owner = "huggingface"
repo = "datasets"

# URL de l'API GitHub pour les issues
url = f"https://api.github.com/repos/{owner}/{repo}/issues"

# Headers pour l'authentification
headers = {
    "Authorization": f"token {token}",
    "Accept": "application/vnd.github.v3+json"
}

# Liste pour stocker les issues
issues_data = []

# Fonction pour récupérer et traiter les issues
def fetch_issues(url, headers, params):
    issues = []
    page = 1
    while True:
        response = requests.get(url, headers=headers, params={**params, "page": page})
        if response.status_code != 200:
            print(f"Failed to fetch issues: {response.status_code}")
            break
        data = response.json()
        if not data:
            break
        issues.extend(data)
        page += 1
    return issues

# Appel de la fonction pour récupérer les issues fermées et ouvertes
states = ["closed", "open"]
for state in states:
    params = {"state": state}
    issues = fetch_issues(url, headers, params)
    
    # Traitement des données d'issues
    for issue in issues:
        issue_number = issue['number']
        issue_title = issue['title']
        issue_labels = [label['name'] for label in issue['labels']]
        issue_user = issue['user']['login']
        issue_state = issue['state']
        created_at = issue['created_at']
        updated_at = issue['updated_at']
        
        # Conversion des dates en objets datetime
        created_date = datetime.strptime(created_at, '%Y-%m-%dT%H:%M:%SZ')
        updated_date = datetime.strptime(updated_at, '%Y-%m-%dT%H:%M:%SZ')
        
        # Calcul de la durée entre la création et la mise à jour
        duration = updated_date - created_date
        
        # Formatage des dates et de la durée
        created_at_formatted = created_date.strftime('%Y-%m-%d %H:%M:%S')
        updated_at_formatted = updated_date.strftime('%Y-%m-%d %H:%M:%S')
        duration_formatted = str(duration)
        
        print(f"Issue #{issue_number}: {issue_title} - Labels: {issue_labels}")
        issues_data.append({
            "number": issue_number,
            "title": issue_title,
            "labels": issue_labels,
            "user": issue_user,
            "state": issue_state,
            "created_at": created_at_formatted,
            "updated_at": updated_at_formatted,
            "duration": duration_formatted
        })

# Affichage des données complètes des issues
for issue in issues_data:
    print(issue)
```

```python
import json

# Sauvegarde des données dans un fichier JSON
with open('issues.json', 'w') as file:
    json.dump(issues_data, file, indent=4)

print("Issues saved to issues.json")
```
